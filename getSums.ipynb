```python
import pandas as pd
```

```python
df = pd.read_csv('Chicago_taxi_trips2017.csv', nrows = 1000000)
```

```python
pd.set_option('display.max_columns', 500)
df.head()
```

```python
df = df.sort_values(by=["Trip Start Timestamp"])
```

```python
df.head()
```

```python
pd.options.display.max_rows = 9999
```

```python
df = df.rename(columns={'Trip Start Timestamp': 'start_timestamp'})
```

```python
df['datetime_values'] = df['start_timestamp'].map(lambda x: pd.to_datetime(x))
```

```python
df['week'] = df['datetime_values'].dt.week
```

```python
import collections
weeks = range(1,54)
trip_miles={}
for week in weeks:
    df_week = df[(df['week'] == week)]
    for _, val in df_week[['Taxi ID', 'Trip Miles']].iterrows():
        if val[0] not in trip_miles:
            trip_miles[val[0]] = [0] * 53
        else:
            trip_miles[val[0]][week - 1] += val[1]
```
